```python
from ultralytics import YOLO
import cv2
import os
```

```python
model = YOLO(
    "/content/drive/MyDrive/RoadShield-AI/models/yolov8_pothole-2/weights/best.pt"
)
```

```python
# Specify Video Path
video_path = "/content/drive/MyDrive/RoadShield-AI/videos/ruralRoad_potHoles.mp4"
```

```python
# Run YOLO on Video
# YOLO automatically:
  # Opens the video
  # Reads every frame
  #  Detects potholes
  #  Draws bounding boxes
  #  Saves the processed video

results = model.predict(
    source=video_path,
    conf=0.25,
    save=True
)
```
